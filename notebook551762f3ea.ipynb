```python
import os
for dirname,_,filenames in os.walk('/kaggle/input'): 
    for filename in filenames: 
        print(os.path.join(dirname, filename))

import pandas as pd
pd.plotting.register_matplotlib_converters() 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

dc_file_path = '../input/fivethirtyeight-comic-characters-dataset/dc-wikia-data.csv'

dc_data = pd.read_csv(dc_file_path, index_col = 'ID')
dc_data = dc_data.drop('GSM', axis=1)

# dropping empty values to give more accurate visuals
dc_data.dropna(axis = 0, inplace=True)
#Set Title

#Setting plot Area Size and background style for plotting
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
plt.title('GENDER APPEARANCES RATIO')

sns.kdeplot(data = dc_data, x = 'APPEARANCES', hue = 'SEX', fill = True)
plt.savefig('myplot.png')

""" it can be deduced from the visuals below that the male and female gender appear almost as much as each other, although the female with relatively lower numbers of characters 
however, the non Genderless characters appearances was way too low"""
```
